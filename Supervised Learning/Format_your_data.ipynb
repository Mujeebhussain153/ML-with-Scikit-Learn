```python
df = pd.DataFrame(data)
df
```

```python
df["is_spam"] = df["label"].map({"spam":1, "not_spam":0})
df[["label", "is_spam"]]
```

```python
# Converting Dummy Data
df = pd.get_dummies(df, columns=["sender_domain"], prefix="domain", dtype="int")
df
```
