```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
X,y = make_moons(n_samples=1000, noise=0.25, random_state=42)
# Selecting all rows and 1st column for X-axis and all rows and 2nd column for the Scatter plot, c means coloring
plt.scatter(X[:,0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolor='k')
plt.title("Make Moons - Scatter Plot")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.show()
```

```python
tree = DecisionTreeClassifier(max_depth=2, random_state=1)
tree.fit(X,y)
```

```python
plt.figure(figsize=(14,6))
plot_tree(tree, filled=True, feature_names=["Feature1", "Feature2"], class_names=["Class 0", "Class 1"])
plt.title("Decision Tree Visualization")
plt.show()
```
